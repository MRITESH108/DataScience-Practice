```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5]
y=[2.3,5,9,8,10]
plt.plot(x,y)
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Line Plot")
plt.plot(x,y,color='red',linestyle='-.')
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle='-.',marker='o' )
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle='-.',marker='o', linewidth=3 )
plt.grid()
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```
