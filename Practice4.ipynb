```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips= sns.load_dataset('tips')
tips
```

```python
# create a scatter plot

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter plot of total bill vs tip")
plt.show()
```

```python
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of size vs total bill")
plt.show()
```

```python
# categorical plot
# bar plot

sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Bar plot of day vs total bill")
plt.show()
```

```python
# box plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title("Box plot of day vs total bill")
plt.show()
```

```python
# violin plot

sns.violinplot(x='day', y='total_bill', data=tips)
plt.title("Violin plot of day vs total bill")
plt.show()
```

```python
# histogram

sns.histplot(x='total_bill', data=tips)
plt.title("Histogram of total bill")
plt.show()
```

```python
sns.histplot(tips['total_bill'],bins=10, kde=True)
plt.title("Histogram of total bill")
plt.show()
```

```python
sns.histplot(tips['total_bill'],bins=100, kde=False)
plt.title("Histogram of total bill")
plt.show()
```

```python
# KDE Plot

sns.kdeplot(tips['total_bill'])
plt.title("KDE Plot of total bill")
plt.show()
```

```python
# KDE Plot

sns.kdeplot(tips['total_bill'],shade=True)
plt.title("KDE Plot of total bill")
plt.show()
```

```python
# pairplot

sns.pairplot(tips)
plt.show()
```

```python
# heatmap

corr= tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')

# plt.show()
```
